## Distribution


### Determine how the data is distributed


The column `ConvertedComp` contains Salary converted to annual USD salaries using the exchange rate on 2019-02-01.

This assumes 12 working months and 50 working weeks.


Plot the distribution curve for the column `ConvertedComp`.


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_salary_distribution(csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Filter out rows where ConvertedComp is not NaN
    df = df[df['ConvertedComp'].notna()]

    # Create a distribution plot using seaborn
    plt.figure(figsize=(10, 6))
    sns.histplot(df['ConvertedComp'], kde=True, bins=20, color='skyblue')
    plt.title('Distribution of Converted Compensation')
    plt.xlabel('Converted Compensation (USD)')
    plt.ylabel('Frequency')
    plt.show()

# Replace 'your_file.csv' with the path to your CSV file
plot_salary_distribution('C:\\Users\\Toshiba\\Downloads\\IBM_final\\m3.csv')
```

## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def box_plot_for_outliers(csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Filter out rows where ConvertedComp is not NaN
    df = df[df['ConvertedComp'].notna()]

    # Create a box plot using seaborn
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df['ConvertedComp'], color='skyblue')
    plt.title('Box Plot of Converted Compensation')
    plt.xlabel('Converted Compensation (USD)')
    plt.show()

# Replace 'your_file.csv' with the path to your CSV file
box_plot_for_outliers('C:\\Users\\Toshiba\\Downloads\\IBM_final\\m3.csv')
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
import pandas as pd

def calculate_iqr(csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Filter out rows where ConvertedComp is not NaN
    df = df[df['ConvertedComp'].notna()]

    # Calculate the first quartile (Q1)
    q1 = df['ConvertedComp'].quantile(0.25)

    # Calculate the third quartile (Q3)
    q3 = df['ConvertedComp'].quantile(0.75)

    # Calculate the Interquartile Range (IQR)
    iqr = q3 - q1

    print(f"The Interquartile Range (IQR) for the 'ConvertedComp' column is: {iqr} USD")

# Replace 'your_file.csv' with the path to your CSV file
calculate_iqr('C:\\Users\\Toshiba\\Downloads\\IBM_final\\m3.csv')
```

Find out the upper and lower bounds.


```python
# your code goes here
import pandas as pd

def calculate_bounds(csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Filter out rows where ConvertedComp is not NaN
    df = df[df['ConvertedComp'].notna()]

    # Calculate the first quartile (Q1)
    q1 = df['ConvertedComp'].quantile(0.25)

    # Calculate the third quartile (Q3)
    q3 = df['ConvertedComp'].quantile(0.75)

    # Calculate the Interquartile Range (IQR)
    iqr = q3 - q1

    # Calculate the upper bound
    upper_bound = q3 + 1.5 * iqr

    # Calculate the lower bound
    lower_bound = q1 - 1.5 * iqr

    print(f"The upper bound for potential outliers is: {upper_bound} USD")
    print(f"The lower bound for potential outliers is: {lower_bound} USD")

# Replace 'your_file.csv' with the path to your CSV file
calculate_bounds('C:\\Users\\Toshiba\\Downloads\\IBM_final\\m3.csv')
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
import pandas as pd

def count_outliers(csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Filter out rows where ConvertedComp is not NaN
    df = df[df['ConvertedComp'].notna()]

    # Calculate the first quartile (Q1)
    q1 = df['ConvertedComp'].quantile(0.25)

    # Calculate the third quartile (Q3)
    q3 = df['ConvertedComp'].quantile(0.75)

    # Calculate the Interquartile Range (IQR)
    iqr = q3 - q1

    # Calculate the upper bound
    upper_bound = q3 + 1.5 * iqr

    # Calculate the lower bound
    lower_bound = q1 - 1.5 * iqr

    # Identify outliers using the bounds
    outliers = df[(df['ConvertedComp'] > upper_bound) | (df['ConvertedComp'] < lower_bound)]

    # Count the number of outliers
    num_outliers = outliers.shape[0]

    print(f"The number of outliers in the 'ConvertedComp' column is: {num_outliers}")

# Replace 'your_file.csv' with the path to your CSV file
count_outliers('C:\\Users\\Toshiba\\Downloads\\IBM_final\\m3.csv')
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
import pandas as pd

def remove_outliers_and_save(csv_file, output_csv):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Filter out rows where ConvertedComp is not NaN
    df = df[df['ConvertedComp'].notna()]

    # Calculate the first quartile (Q1)
    q1 = df['ConvertedComp'].quantile(0.25)

    # Calculate the third quartile (Q3)
    q3 = df['ConvertedComp'].quantile(0.75)

    # Calculate the Interquartile Range (IQR)
    iqr = q3 - q1

    # Calculate the upper bound
    upper_bound = q3 + 1.5 * iqr

    # Calculate the lower bound
    lower_bound = q1 - 1.5 * iqr

    # Create a new DataFrame by removing outliers
    df_no_outliers = df[(df['ConvertedComp'] <= upper_bound) & (df['ConvertedComp'] >= lower_bound)]

    # Save the new DataFrame to a new CSV file
    df_no_outliers.to_csv(output_csv, index=False)

    print(f"New DataFrame without outliers saved to '{output_csv}'.")

# Replace 'your_file.csv' with the path to your CSV file
# Replace 'output_no_outliers.csv' with the desired name for the new CSV file
remove_outliers_and_save('C:\\Users\\Toshiba\\Downloads\\IBM_final\\m3.csv', 'm3p.csv')


   
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.


```python
# your code goes here
import pandas as pd

def find_correlation(csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Filter out rows where Age is not NaN
    df_age = df[df['Age'].notna()]

    # Find the correlation between Age and all other numerical columns
    correlation_matrix = df_age.corrwith(df_age['Age'])

    print("Correlation between 'Age' and other numerical columns:")
    print(correlation_matrix)

# Replace 'your_file.csv' with the path to your CSV file
find_correlation('C:\\Users\\Toshiba\\Downloads\\IBM_final\\m3p.csv')
```
