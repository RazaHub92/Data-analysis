## Removing duplicates


Remove the duplicate rows from the dataframe.


```python
# your code goes here
import pandas as pd

def remove_duplicate_rows(csv_file, output_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Remove duplicate rows and update the DataFrame
    df_no_duplicates = df.drop_duplicates()

    # Save the DataFrame without duplicates to a new CSV file
    df_no_duplicates.to_csv(output_file, index=False)

    print(f"Duplicate rows removed. Updated CSV saved to {output_file}")

# Replace 'your_input_file.csv' and 'your_output_file.csv' with the actual file paths
remove_duplicate_rows('C:\\Users\\Toshiba\\Downloads\\IBM_final\\m1.csv', 'C:\\Users\\Toshiba\\Downloads\\IBM_final\\m2.csv')
```

## Finding Missing values


Find the missing values for all columns.


```python
# your code goes here
import pandas as pd

def find_missing_values(csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Identify missing values in each column
    missing_values = df.isnull().sum()

    # Display the columns with missing values and their counts
    print("Missing Values:")
    print(missing_values[missing_values > 0])

# Replace 'your_file.csv' with the path to your CSV file
find_missing_values('C:\\Users\\Toshiba\\Downloads\\IBM_final\\m2.csv')
```

## Imputing missing values


Find the  value counts for the column WorkLoc.


```python
# your code goes here
import pandas as pd

def get_value_counts(csv_file, column_name):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Get the value counts for the specified column
    value_counts = df[column_name].value_counts()

    print(f"Value Counts for the '{column_name}' column:")
    print(value_counts)

# Replace 'your_file.csv' with the path to your CSV file
# Replace 'WorkLoc' with the name of the column you're interested in
get_value_counts('C:\\Users\\Toshiba\\Downloads\\IBM_final\\m2.csv', 'UndergradMajor')
```

Identify the value that is most frequent (majority) in the WorkLoc column.


```python
#make a note of the majority value here, for future reference
import pandas as pd

def find_most_frequent_value(csv_file, column_name):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Find the most frequent value in the specified column
    most_frequent_value = df[column_name].mode().iloc[0]

    print(f"The most frequent value in the '{column_name}' column is: {most_frequent_value}")

# Replace 'your_file.csv' with the path to your CSV file
# Replace 'WorkLoc' with the name of the column you're interested in
find_most_frequent_value('C:\\Users\\Toshiba\\Downloads\\IBM_final\\m2.csv', 'WorkLoc')
```

Impute (replace) all the empty rows in the column WorkLoc with the value that you have identified as majority.


```python
import os
import pandas as pd

def impute_missing_values(csv_file, column_name, impute_value):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Impute missing values in the specified column with the provided value
    df[column_name] = df[column_name].fillna(impute_value)

    # Create a directory to save the imputed file if it doesn't exist
    output_directory = 'imputed_data'
    os.makedirs(output_directory, exist_ok=True)

    # Save the DataFrame with imputed values to a new CSV file in the created directory
    output_file = os.path.join(output_directory, 'imputed_' + os.path.basename(csv_file))
    df.to_csv(output_file, index=False)

    print(f"Missing values in the '{column_name}' column imputed with '{impute_value}'.")
    print(f"Updated CSV saved to '{output_file}'.")

# Replace 'your_file.csv' with the path to your CSV file
# Replace 'WorkLoc' with the name of the column you're interested in
# 'Office' is the value to impute with empty rows in the 'WorkLoc' column


impute_missing_values('C:\\Users\\Toshiba\\Downloads\\IBM_final\\m2.csv', 'WorkLoc', 'Office')
```

After imputation there should ideally not be any empty rows in the WorkLoc column.


Verify if imputing was successful.


```python
# your code goes here
import pandas as pd

def verify_imputation(csv_file, column_name):
    # Read the imputed CSV file into a pandas DataFrame
    df_imputed = pd.read_csv(csv_file)

    # Check for missing values in the specified column
    missing_values_after_imputation = df_imputed[column_name].isnull().sum()

    if missing_values_after_imputation == 0:
        print(f"Imputation was successful. No missing values in the '{column_name}' column.")
    else:
        print(f"Imputation was not successful. There are still {missing_values_after_imputation} missing values in the '{column_name}' column.")

# Replace 'imputed_your_file.csv' with the path to your imputed CSV file
# Replace 'WorkLoc' with the name of the column you're interested in
verify_imputation('C:\\Users\\Toshiba\\Downloads\\IBM_final\\imputed_m2.csv', 'WorkLoc')
```

## Normalizing data


There are two columns in the dataset that talk about compensation.

One is "CompFreq". This column shows how often a developer is paid (Yearly, Monthly, Weekly).

The other is "CompTotal". This column talks about how much the developer is paid per Year, Month, or Week depending upon his/her "CompFreq". 

This makes it difficult to compare the total compensation of the developers.

In this section you will create a new column called 'NormalizedAnnualCompensation' which contains the 'Annual Compensation' irrespective of the 'CompFreq'.

Once this column is ready, it makes comparison of salaries easy.


<hr>


List out the various categories in the column 'CompFreq'


```python
# your code goes here
import pandas as pd

def list_compensation_frequencies(csv_file, column_name):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Get the unique values in the specified column
    comp_freq_categories = df[column_name].unique()

    print(f"Categories in the '{column_name}' column:")
    for category in comp_freq_categories:
        print(category)

# Replace 'your_file.csv' with the path to your CSV file
# Replace 'CompFreq' with the name of the column you're interested in
list_compensation_frequencies('C:\\Users\\Toshiba\\Downloads\\IBM_final\\imputed_m2.csv', 'CompFreq')
```

Create a new column named 'NormalizedAnnualCompensation'. Use the hint given below if needed.


Double click to see the **Hint**.

<!--

Use the below logic to arrive at the values for the column NormalizedAnnualCompensation.

If the CompFreq is Yearly then use the exising value in CompTotal
If the CompFreq is Monthly then multiply the value in CompTotal with 12 (months in an year)
If the CompFreq is Weekly then multiply the value in CompTotal with 52 (weeks in an year)

-->


```python
# your code goes here
import os
import pandas as pd

def create_normalized_compensation_column(csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Define a function to apply the normalization logic
    def normalize_compensation(row):
        if row['CompFreq'] == 'Yearly':
            return row['CompTotal']
        elif row['CompFreq'] == 'Monthly':
            return row['CompTotal'] * 12
        elif row['CompFreq'] == 'Weekly':
            return row['CompTotal'] * 52
        else:
            return None  # Handle other cases or missing values as needed

    # Create the 'NormalizedAnnualCompensation' column using the normalization function
    df['NormalizedAnnualCompensation'] = df.apply(normalize_compensation, axis=1)

    # Create a directory to save the file if it doesn't exist
    output_directory = 'normalized_data'
    os.makedirs(output_directory, exist_ok=True)

    # Save the DataFrame with the new column to a new CSV file in the created directory
    output_file = os.path.join(output_directory, 'normalized_' + os.path.basename(csv_file))
    df.to_csv(output_file, index=False)

    print(f"New column 'NormalizedAnnualCompensation' created and saved to '{output_file}'.")
# Replace 'your_file.csv' with the path to your CSV file
create_normalized_compensation_column('C:\\Users\\Toshiba\\Downloads\\IBM_final\\imputed_m2.csv')
```
